### Reading the datafile for drug discovery with the help of read_csv file of Pandas library

```python
import pandas as pd

df = pd.read_csv("refined_data_heatmap.csv")

df.head()
```

### Exploratory Data Analysis using Heatmap of cross-correlation matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.gcf().set_size_inches(15, 15)

cmap=sns.diverging_palette(500,10,as_cmap=True)

sns.heatmap(df.corr(),cmap=cmap,center=0,annot=False,square=True)

plt.show()
```
