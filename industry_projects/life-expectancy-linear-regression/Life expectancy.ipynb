### We will be mapping the countries to their respective continents and plotting the bubble plot for every continent

```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

from pycountry_convert import country_alpha2_to_continent_code as to_continent
from pycountry_convert import country_name_to_country_alpha2 as to_alpha

continents = {
    'NA': 'North America',
    'SA': 'South America', 
    'AS': 'Asia',
    'OC': 'Australia',
    'AF': 'Africa',
    'EU': 'Europe'
}

continent = []
for country in life_exp['Country']:
    try:
        continent.append(continents[(to_continent(to_alpha(country)))])
    except:
        continent.append("Africa")

        
life_exp["Continent"] = continent
to_bubble = life_exp[["Country", "Year", "Life expectancy ", "GDP", "Population", "Continent"]]
to_bubble.dropna(inplace = True)
```

```python
df = px.data.gapminder()

fig = px.scatter(to_bubble, x="GDP", y="Life expectancy ",
                 size="Population", color="Continent",
                 hover_name="Country", log_x=True, size_max=50)
fig.show()
```

### More analysis on the dataset to check the effect of 'Income composition of resources' with respect to life expectancy for every continent

```python
fig, axs = plt.subplots(2,3, figsize=(15, 6), facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace = 0.5)

for continent, ax in zip(set(life_exp["Continent"]), axs.flat):
    
    continents = life_exp[life_exp['Continent'] == continent]
    
    sns.regplot(x = continents['Income composition of resources'],y = continents["Life expectancy "], color = 'blue', ax = ax).set_title(continent)

plt.tight_layout()    
plt.show()
```

### Now we will observe how all features are affecting the life-expectancy in a certain continent

```python
nr_rows = 5
nr_cols = 3

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*5,nr_rows*5))
numerical_feats = life_exp.dtypes[life_exp.dtypes != "object"].index
li_num_feats = list(numerical_feats)
li_not_plot = []
li_plot_num_feats = [c for c in list(numerical_feats) if c not in li_not_plot]
life_exp.dropna(inplace = True)

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        i = r*nr_cols+c
        if i < len(li_plot_num_feats):
            sns.regplot(x = life_exp[li_plot_num_feats[i]],y = life_exp["Life expectancy "] , color = 'blue', ax = axs[r][c])
            stp = stats.pearsonr(life_exp[li_plot_num_feats[i]], life_exp["Life expectancy "])
            str_title = "r = " + "{0:.2f}".format(stp[0]) + "      " "p = " + "{0:.2f}".format(stp[1])
            axs[r][c].set_title(str_title,fontsize=11)
            
            
plt.tight_layout()    
sns.set(color_codes=True)
plt.show();
```

### Segregating target variable from the input features

```python
target = life_exp['Life expectancy ']
features = life_exp[life_exp.columns.difference(['Life expectancy ','Year'])]
```

### Splitting the dataset into training and testing sets using the scikit-learns' train_test_split function in the ratio of 70:30.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(pd.get_dummies(features), target, test_size=0.3)
```

### Fitting the [linear regression](https://www.enjoyalgorithms.com/blog/linear-regression-in-machine-learning/) model the train dataset

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

### Finding the confidence score in the LR fitting

```python
lr_confidence = lr.score(X_test, Y_test)
print("lr confidence: ", lr_confidence)
```

### Plotting the model predictions' residuals to check if the fitting is perfect?

```python
model_predictions = lr.predict(X_test)
residuals = model_predictions - Y_test
```

```python
plt.figure(figsize=(15,5))
sns.histplot(residuals, kde=True, color="orange")
plt.title('Residual Plot')
plt.xlabel('Residuals: (Predictions - Actual)')
plt.ylabel('Density');
```

### Fitting the different Linear Regression model using the statsmodel library and Oridinary Least Square (OLS) method

```python
import statsmodels.api as sm

X_addC = sm.add_constant(X_train)

result = sm.OLS(Y_train, X_addC).fit()

print(result.rsquared, result.rsquared_adj)
```
